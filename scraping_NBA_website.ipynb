I am going to start extracting data from the NBA website and see what it looks like.

```python
import nba_scraper.nba_scraper as ns

# playing with game IDs

nba_df = ns.scrape_game([21800001, 21800002])
nba_df.head()
```

It looks like with `scrape_game` I obtain the chronological series of events in a game and who was involved.

It is not easy to determine the correspondence between a game ID and the game details (season, date, teams, etc).

Let's see what information can we get from a whole season.

```python
#scrape a season
nbaseas_df = ns.scrape_season(2019)
head(nbaseas_df)
```

It seems that `nba_scraper` is very limited: only game, season or date interval stats can be accessed. No player or team based search utilities.
